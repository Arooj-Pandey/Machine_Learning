```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
#here "plt.xlabel() is used to assign the name of the x axis" similarly "ylabel is used to assign the name of the yaxis"
plt.xlabel("days")
plt.ylabel("temperature")
plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")
plt.legend( loc = "lower left", shadow = True)

plt.grid()

# label in plot function is used to label the lines of the graph and "using LEGEND() is must to execute the statement"

#(loc="best") attribute can be used to decide the best location of the label  

# plt.grid() makes it easy to read the graph 
```
