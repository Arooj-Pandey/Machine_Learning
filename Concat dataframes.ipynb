```python
import pandas as pd
india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature":[32,45,30],
    "Humidity":[80,68,78]
})
india_weather
#pd.DataFrame should be written in the way it is only, if written in smaller case it may give error
```

```python
us_weather = pd.DataFrame({
    "city": ["NewYork","Chicago","Orlando"],
    "temperature":[21,14,35],
    "Humidity":[68,65,75]
})
us_weather
```

```python
#to merge the above tables into one we use "concat function" , you can also use it to merge more than 2 dataframes
df=pd.concat([india_weather,us_weather])
df
```

```python
#in the above table as we can see concat is still using the original indices of both data frames, so in order to get continuous 
#index in our newly merged table we use argument "ignore_index"
df=pd.concat([india_weather,us_weather],ignore_index = True)
df
```

```python
#using "keys=[]" argument in concaat we can assign a key to the merged tables in order to get a more clearer view about what 
#belongs to which table, but it cannot be used with ignore_index argument
df=pd.concat([india_weather,us_weather],keys=["india","us"])
df
```

```python
#we can also call the subset of the merged data set which belongs to the key passed in "loc[]" function
df.loc["india"]
```

```python
df.loc["us"]
```

```python
temperature_df = pd.DataFrame({
    "city": ["NewYork","Chicago","Orlando"],
    "temperature":[21,14,35],
    
})
temperature_df
```

```python
wind_speed_df = pd.DataFrame({
    "city": ["NewYork","Chicago","Orlando"],
    "windspeed":[7,12,9]
} )
wind_speed_df
```

```python
#if you want a final data frame which consists of city,temperature and wind_speed, then to append the following columns in the 
# original dataframe we use
df=pd.concat([temperature_df,wind_speed_df])
df 
#here we didn't need any additional rows, we just wanted them to get added as columns, To do so we use "axis = 1" argument
```

```python
df=pd.concat([temperature_df,wind_speed_df], axis = 1)
df
```
