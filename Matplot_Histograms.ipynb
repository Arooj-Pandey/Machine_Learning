```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar=[113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar, bins=3)
#here in histogram we only need to pass a 1-D array and the frequency will be addded automatically
#using "bins" we can control the number of bars in our histograms. 
#Default "bin" parameter is 10
```

```python
blood_sugar=[113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar, bins=3, rwidth=0.98, color="red")
#using "rwidth" you can control the spaces b/w the graphs
```

```python
blood_sugar=[113,85,90,150,149,88,93,115,135,80,77,82,129]
plt.hist(blood_sugar, bins=3, rwidth=0.98, color="red", histtype='step')
```

```python
blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82]
blood_sugar_women=[67,98,89,120,133,155,84,69,89,79,120,112,100]

plt.xlabel('sugar range')
plt.ylabel('Total number of patients')
plt.title('Blood Sugar Analysis')

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.98, color=['grey','yellow'],
         label=['men','women'])

plt.legend()
```

```python
plt.xlabel('sugar range')
plt.ylabel('Total number of patients')
plt.title('Blood Sugar Analysis')

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.98, color=['grey','yellow'],
         label=['men','women'], orientation='horizontal')

plt.legend()
```
