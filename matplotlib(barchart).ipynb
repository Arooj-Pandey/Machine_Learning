```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company=['GOOGL', 'AMZN', 'MSFT', 'FB']
revenue=[90,136,89,27]
profit=[40,19,30,5]

#here we can not just plot a graph based on plt.bar(company,revenue) it will simply create an error since company contains strings
```

```python
ypos = np.arange(len(company))
ypos

#therefore we used ypos of numpy module to create an array of the company named "ypos"
```

```python
plt.bar(ypos,revenue)
```

```python
plt.title("US TECH STOCKS")

plt.xlabel("COMPANY")

plt.ylabel("Revenue(USD)")

plt.xticks(ypos , company)

plt.bar(ypos,revenue, label = "revenue")

plt.legend()
# Now here the corresponding elements of the array "ypos" were replaced with there actual string value of company using "xticks()"
```

```python
plt.title("US TECH STOCKS")

plt.xlabel("COMPANY")

plt.ylabel("Revenue(USD)")

plt.xticks(ypos , company)

plt.bar(ypos,revenue, label = "revenue")
plt.bar(ypos,profit, label ="profit")

plt.legend()

#here we plotted the profit chart as well in the same chat, but we need the profi bar side by side to the revenue bar
```

```python
plt.title("US TECH STOCKS")

plt.xlabel("COMPANY")

plt.ylabel("Revenue(USD)")

plt.xticks(ypos , company)

plt.bar(ypos-0.2,revenue,width=0.4, label = "revenue")
plt.bar(ypos+0.2,profit,width=0.4, label ="profit")

plt.legend()

#only by positioning the two bars differently and working up with their widths, we had our perfect "Profit-Revenue" graph
```

```python
plt.title("US TECH STOCKS")

plt.ylabel("COMPANY")

plt.xlabel("Revenue(USD)")

plt.yticks(ypos , company)

plt.barh(ypos-0.2,revenue, label = "revenue")
plt.barh(ypos+0.2,profit, label ="profit")

plt.legend()

# plt.barh is used for horizontal bar
```
