```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("C:\\Users\\arooj\\Downloads\\canada_per_capita_income.csv")
df.head(5)
```

```python
%matplotlib inline 
plt.xlabel('per capita income (US$')
plt.ylabel('year',fontsize='15')
plt.scatter(df['per capita income (US$)'], df['year'], color='red',marker='+')
```

```python
#it's important that what you put first in the "fit()" function, also the first value always has to be in a 2-D array followed by
 
#the second value which should be declared in a 1-D array or simply as "df.per capita income" but not in this case....why
rg=linear_model.LinearRegression()
rg.fit(df[['year']],df['per capita income (US$)'])

```

```python
rg.intercept_
```

```python
rg.predict([[2020]])
```

```python
rg.predict([[2025]])
```

```python
%matplotlib inline
plt.xlabel('per capita income', fontsize=20)
plt.ylabel('year',fontsize = 20)
plt.scatter(df['year'] , df['per capita income (US$)'], color = 'blue', marker='+')
plt.plot(df['year'], rg.predict(df[['year']]), color ='purple',linestyle='dotted')


```
