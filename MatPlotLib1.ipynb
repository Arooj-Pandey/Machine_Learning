```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
# 'x and y' are python lists where x contains number of days while y contains temp. associated with them
```

```python
plt.plot(x,y, color='purple', linewidth='2', linestyle='dotted', alpha=0.5)
#here alpha is used to manage the opacity of the line
```

```python
plt.plot(x,y,alpha=0.8,marker='D',linestyle="")
```
